## Identify Good Candidates for Log Transformation

Below we plot each of the potential numeric features against `SalePrice`:

```python
# Run this cell without changes

y = ames["SalePrice"]
X = ames.drop("SalePrice", axis=1)

fig, axes = plt.subplots(nrows=6, ncols=6, figsize=(30, 20), sharey=True)

for i, column in enumerate(X.columns):
    # Locate applicable axes
    row = i // 6
    col = i % 6
    ax = axes[row][col]
    
    # Plot feature vs. y and label axes
    ax.scatter(X[column], y, alpha=0.2)
    ax.set_xlabel(column)
    if col == 0:
        ax.set_ylabel("SalePrice")

plt.show()
fig.tight_layout()
```

### Plot Log Transformed Versions of Features

For each feature that you identified as a good candidate for log transformation, plot the feature vs. `SalePrice` as well as the log transformed feature vs. log transformed `SalePrice`.

```python
X_log = ames.copy()
X_log['log(GarageArea)'] = np.log(X_log['GarageArea'])

y_log = np.log(y)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 5))

ax1.scatter(ames['GarageArea'], y)
ax1.set_xlabel('GarageArea')
ax1.set_ylabel('SalePrice')

ax2.scatter(X_log['log(GarageArea)'], y_log, color='red')
ax2.set_xlabel('log(GarageArea)')
ax2.set_ylabel('log(SalePrice)')

plt.show()
```

```python
X_log['log(GrLivArea)'] = np.log(X_log['GrLivArea'])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 5))

ax1.scatter(ames['GrLivArea'], y)
ax1.set_xlabel('GrLivArea')
ax1.set_ylabel('SalePrice')

ax2.scatter(X_log['log(GrLivArea)'], y_log, color='red')
ax2.set_xlabel('log(GrLivArea)')
ax2.set_ylabel('log(SalePrice)')

plt.show()
```

```python
X_log['log(1stFlrSF)'] = np.log(X_log['1stFlrSF'])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 5))

ax1.scatter(ames['1stFlrSF'], y)
ax1.set_xlabel('1stFlrSF')
ax1.set_ylabel('SalePrice')

ax2.scatter(X_log['log(1stFlrSF)'], y_log, color='red')
ax2.set_xlabel('log(1stFlrSF)')
ax2.set_ylabel('log(SalePrice)')

plt.show()
```

## Build a Model with Log-Transformed Features and Target

Choose up to 3 of the features you investigated, and set up an X dataframe containing the log-transformed versions of these features as well as a y series containing the log-transformed version of the target.

```python
# Your code here - prepare data for modeling
transformed_X = X_log[['log(GrLivArea)', 'log(GarageArea)', 'log(1stFlrSF)']]

transformed_X
```

```python
raw_X = X[['GrLivArea', 'GarageArea', '1stFlrSF']]
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model
model = sm.OLS(y_log, sm.add_constant(transformed_X))
model_results = model.fit()

print(model_results.summary())
```

```python
# untransformed features and target just for comparison
model_1 = sm.OLS(y, sm.add_constant(raw_X))
model1_results = model_1.fit()

print(model1_results.summary())
```

### Model Evaluation and Interpretation

How did the model perform? How might we interpret its coefficients? Create as many cells as needed.

```python
# Your code here - evaluate the model
model_results.params
```

```python
print(f'GrLivArea Beta: {(np.exp(np.log(1.01) * model_results.params['log(GrLivArea)'])) - 1}')
print(f'GarageArea Beta: {(np.exp(np.log(1.01) * model_results.params['log(GarageArea)'])) - 1}')
print(f'1stFlrSF Beta: {(np.exp(np.log(1.01) * model_results.params['log(1stFlrSF)'])) - 1}')
```

The model is statistically significant and explains about 65.5% of the variance in sales price which means other important factors that influence the sale price haven't being included.
